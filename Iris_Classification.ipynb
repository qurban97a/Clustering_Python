```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_scaled=sc.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pc = PCA(svd_solver='auto', n_components =2)
afterPcaData=pd.DataFrame(pc.fit_transform(data_scaled))
print('Total no. of principal components =',pc.n_components_)
```

```python
afterPcaData
```

```python
fig = plt.figure(figsize = (12,6))
sns.scatterplot( afterPcaData[0],afterPcaData[1],  s=400)
plt.title('PC1 vs PC2',fontsize=15)
plt.xlabel('Principal Component 1',fontsize=12)
plt.ylabel('Principal Component 2',fontsize=12)
plt.show();
```

## Clustering- KMeans

```python
from sklearn.cluster import KMeans
a=[]
K=range(1,10)
for i in K:
    kmean=KMeans(n_clusters=i)
    kmean.fit(afterPcaData)
    a.append(kmean.inertia_)
    
plt.plot(K,a,marker='o')
plt.title('Elbow Method',fontsize=15)
plt.xlabel('Number of clusters',fontsize=15)
plt.ylabel('Sum of Squared distance',fontsize=15)
plt.show()
```

```python
km_3=KMeans(n_clusters=3,init = 'k-means++',random_state=78)
km_3.fit(afterPcaData)
y_pred=km_3.predict(afterPcaData)
y_pred
```

```python
pd.Series(y_pred).value_counts() #Actual count of  spicies is equal to 50 for each classs 
#However predicted counts are mentioned above
```

```python
fig = plt.figure(figsize = (15,8))
plt.scatter(afterPcaData[1],afterPcaData[0],c =y_pred, s=100,label='Standard')
plt.title('Cluster of Species [after PCA]')
plt.show()
```

## Hierarhical Clustering

```python
import scipy.cluster.hierarchy as hcluster

plt.figure(figsize=(50, 12))
dend=hcluster.dendrogram(hcluster.linkage(afterPcaData, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

hcluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
hcluster.fit_predict(afterPcaData)
hcluster_label = hcluster.labels_
```
